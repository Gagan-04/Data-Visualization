# Let us visualize the Performance of Students  

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
per = pd.read_csv("Performance.csv")
```

```python
per.head(50)
```

```python
per.isnull().sum()
```

```python
per.describe()
```

```python
p = sns.countplot(x = "math score", data = per, palette = "muted")
```

```python
passmark = 40
per["mathpass"] = np.where(per["math score"] < passmark, "f", "p")
```

```python
per.mathpass.value_counts()
```

```python
p = sns.countplot(x = "parental level of education", data = per, hue = "mathpass")
a = plt.setp(p.get_xticklabels(), rotation=90)
```

```python
passmark = 40
per["readpass"] = np.where(per["reading score"] < passmark, "f", "p")
```

```python
per.readpass.value_counts()
```

```python
p = sns.countplot(x = "parental level of education", data = per, hue = "readpass")
a = plt.setp(p.get_xticklabels(), rotation=90)
```

```python
passmark = 40
per["writescr"] = np.where(per["writing score"] < passmark, "f", "p")
```

```python
per.writescr.value_counts()
```

```python
p = sns.countplot(x = "parental level of education", data = per, hue = "writescr", palette = "bright")
a = plt.setp(p.get_xticklabels(), rotation = 90)
```
